```python
import random
import numpy as np
```

```python
def sacapelota(Pelotas,CantN,CantU):
    recordList, nueva, usada = [], 0, 0 
    for i in range(Pelotas): 
         flip = random.random()
         #print(flip)
         if (flip <= CantN/(CantN+CantU)):
              #print("nueva")
              recordList.append("nueva")
              CantN = CantN-1;
              nueva = nueva + 1;
         else:
              #print("usada")
              recordList.append("usada")
              CantU = CantU-1;
      
    #print(str(recordList))
    #print(str(recordList.count("nueva")) )
    #print(str(recordList.count("usada")) )
    return nueva
```

```python
salieron2N = 0;
iteraciones =10000;
for i in range(iteraciones):
  #print("Primer partido");
  nuevas = sacapelota(2,15,5)
  #print("Segundo partido");
  if nuevas == 0:
    nuevas2 = sacapelota(2,15,5)
  elif nuevas == 1:
    nuevas2 = sacapelota(2,14,6)  
  else :
    nuevas2 = sacapelota(2,13,7) 
  if nuevas2 == 2:
    salieron2N = salieron2N+1;

print(salieron2N/iteraciones)
```
